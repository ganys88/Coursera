```python
# create 'Month Joined' column
companies['Month Joined'] = companies['Date Joined'].dt.month_name()

# by extracting the month_name from each value in 'Date Joined' we can create 'Month Joined' column
```

```python
# create a 'Years to Join' column
companies['Years to Join'] = companies['Date Joined'].dt.year - companies['Year Founded']

# simple math
# the year it became unicorn company (-) subtract 'Year Founded'.
# given that the 'Date Joined' columns is in datetime format, extract the year component from 'Date Joined' and subtract the 'Year Founded' comuln from it.
```

```python
# gain more insight on a specific year
unicorn_2021 = companies[companies['Date Joined'].dt.year == 2021]

# Filter dataset by a year of your interest (in terms of when companies reached unicorn status).
# Save the resulting subset in a new variable.
```

```python
# observe trends over time
# After identifying the time interval that interests you, proceed with the following:
# Step 1. Take the subset that you defined for the year of interest.
#         Insert a column that contains the time interval that each data point belongs to, as needed.
# Step 2. Group by the time interval.
#         Aggregate by counting companies that joined per interval of that year.
#         Save the resulting DataFrame in a new variable.
```

```python
# insert a 'Week Joined' column into 'unicorn_2021'
unicorn_2021.insert(3, 'Week Joined', unicorn_2021['Date Joined'].dt.strftime('%Y-W%V'), True)
```

```python
# check and confirm
unicorn_2021.head()
```

```python
# now group 'unicorn_2021' by 'Week Joined'
# aggregate by counting companies that joined per week of 2021
# save the resulting dataframe in a new variable

unicorn_by_week_2021 = unicorn_2021.groupby(by='Week Joined')['Company'].count().reset_index().rename(columns={'Company':'Company Count'})
```

```python
# check and confirm
unicorn_by_week_2021.head()
```

```python
# compare trends over time
# After identifying the additional year and time interval of interest, proceed with the following:
# Step 1. Filter by the additional year to create a subset that consists of companies that joined in that year.
# Step 2. Concatenate that new subset with the subset that you defined previously.
# Step 3. As needed, add a column that contains the time interval that each data point belongs to,
#         in the concatenated DataFrame.
# Step 4. Transform the `Valuation` column as needed.
# Step 5. Group by the time interval.
#         Aggregate by computing average valuation of companies that joined per interval of the corresponding year.
#         Save the resulting DataFrame in a new variable.
```

```python
# 1. Filter by the additional year to create a subset that consists of companies that joined in that year.
unicorn_2020 = companies[companies['Date Joined'].dt.year == 2020]
```

```python
# 2. Concatenate the new subset with the subset that you defined previously.
unicorn_2020_2021 = pd.concat([unicorn_2020, unicorn_2021.drop(columns='Week Joined')])
```

```python
# check and confirm
unicorn_2020_2021.head(2)
```

```python
# 3. Add `Quarter Joined` column to `companies_2021`.
unicorn_2020_2021['Quarter Joined'] = unicorn_2020_2021['Date Joined'].dt.to_period('Q').dt.strftime('%Y-Q%q')
```

```python
# check and confirm
unicorn_2020_2021.head(2)
```

```python
# 4. Convert the `Valuation` column to numeric by removing `$` and `B` and casting each value to data type `float`.
unicorn_2020_2021['Valuation'] = unicorn_2020_2021['Valuation'].str.strip('$B').astype(float)

# the strip() function in Python is a string method that removes any leading(at begginning) or trailing(at end) characters from a string including whitespaces.
# the astype() method returns a new dataframe where the dataframe has been changed to specified dtype() in this case (float).
```

```python
# check and confirm
unicorn_2020_2021.head(2)

# $(leading) and B (trailing) has been removed and the value has been changed to float. for eg: $15B to 15.0
```

```python
# 5. # Group `companies_2020_2021` by `Quarter Joined`,
     # Aggregate by computing average `Funding` of companies that joined per quarter of each year.
     # Save the resulting DataFrame in a new variable.

unicorn_by_quarter_2020_2021 = unicorn_2020_2021.groupby(by='Quarter Joined')['Valuation'].mean().reset_index().rename(columns={'Valuation':'Averange Valuation'})
```

```python
# check and confirm
unicorn_by_quarter_2020_2021
```

```python
# time-to-unicorn visualization
# visualize the time it took companies to become unicorn
# using companies dataset create a boxplot

companies.head()
```

```python
# define a list that contains months in chronogical order
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']
print('month order')
print(month_order)
```

```python
# create a box plot to visualize the distribution of how long it took companies to become unicorn
# make sure the x-axis goes in chronological order by month
# plot the data from companies dataframe

sns.boxplot(x=companies['Month Joined'], y=companies['Years to Join'],
            order = month_order,
            showfliers=False)

plt.title('Distribution of years to become Unicorn with respect to month joined')

plt.xticks(rotation=45, horizontalalignment='right')

plt.show()
```

```python
# result and evaluation
# visualize the time it took companies to reach unicorn status
```

```python
# set the size of the plot
plt.figure(figsize=(16,8))

# create a barplot to visualize the average number of years it took companies to reach unicorn status with respect to when they were founded
# plot data from 'companies dataframe
sns.barplot(x=companies['Year Founded'], y=companies['Years to Join'])

plt.xticks(rotation=45, horizontalalignment='right')

plt.title('average number of years to become Unicorn')

plt.show()
```

```python
# visualize the number of companies that joined per interval
# set the size of the plot
plt.figure(figsize=(16,5))
# create a barplot to visualize the number of companies that joined per interval fo the year of interest.
sns.barplot(x=unicorn_by_week_2021['Week Joined'], y=unicorn_by_week_2021['Company Count'])

plt.xticks(rotation=45, horizontalalignment='right')

plt.xlabel('Week Number')
plt.ylabel('Number of Companies')
plt.title('Number of Companies that became unicorn per week in 2021')

plt.show()
```

```python
# visualize the average valuation over the quarters
# using slicing, extract the year component and the time interval that you specified
# and save tehm by adding two new columns in into the subset

unicorn_by_quarter_2020_2021['Quarter Number'] = unicorn_by_quarter_2020_2021['Quarter Joined'].str[-2:]
unicorn_by_quarter_2020_2021['Year Joined'] = unicorn_by_quarter_2020_2021['Quarter Joined'].str[:4]

# check and confirm
unicorn_by_quarter_2020_2021
```

```python
# now vizualize
plt.figure(figsize=(10,5))

sns.barplot(x=unicorn_by_quarter_2020_2021['Quarter Number'], y=unicorn_by_quarter_2020_2021['Averange Valuation'],
            hue=unicorn_by_quarter_2020_2021['Year Joined'])

# hue is used to differentiate categories within a data

plt.ylabel('Average Valuation in Billions')

plt.show()
```
