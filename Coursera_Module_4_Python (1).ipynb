```python
def purchases_100(sales):
    count = 0                           # Set a counter of totals > 100
    for customer in sales:              # Loop over each inner list
        customer_total = 0              # Set 0 value of purchases for the inner list
        for purchase in customer:       # For price in inner list:
            customer_total += purchase  # Add the price to value of purchases for inner list
        if customer_total >= 100:       # After looping over all prices in inner list, if
                                        # total >= 100
            count+=1                    # Increment the counter
    return count
```

```python
def factorial(n):
    result = 1                 # Initialize the result variable to 1
    for i in range(1, n + 1):  # Loop through numbers from 1 to n (inclusive)     .it creates a range, which is from (1,5). starts from 1 and ends to 5 which is 1,2,3,4.
        result *= i            # Multiply the current value of result by the current value of i
    return result              # Return the final result after the loop

result = factorial(4)          # Call the factorial function with argument 4
print(result)                  # Print the result
```

Explanation:

result = 1: Before starting the loop, we initialize a variable result to 1. This variable will store the factorial of the input number n.

for i in range(1, n + 1): This line sets up a loop that iterates over the numbers from 1 to n (inclusive). In this case, n is 4, so the loop will iterate over the numbers 1, 2, 3, and 4.

result *= i: Inside the loop, we multiply the current value of result by the current value of i. This accumulates the product of all numbers from 1 to n into the result variable. In each iteration, i takes on the values 1, 2, 3, and 4, so we are effectively multiplying result by each of these numbers.

After the loop finishes, we have multiplied all the numbers from 1 to n together, resulting in the factorial of n, which is stored in the result variable.

return result: Finally, we return the computed factorial value stored in the result variable.

So, for factorial(4), the loop calculates 1 * 1 * 2 * 3 * 4, which equals 24, and this value is returned.

# **Boolean masking**
A filtering technique that overlays a Boolean grid onto a dataframe in order to select only the values in the dataframe that align with the True values of the grid.

#   Operator Logic


*  ampersand  =  &  ,   and     
*  verical bar =  |   ,  or     
*   tilda =  ~  ,   not     


```python
data = {'planet': ['Mercury', 'Venus', 'Earth', 'Mars',
                   'Jupiter', 'Saturn', 'Uranus', 'Neptune'],
       'radius_km': [2440, 6052, 6371, 3390, 69911, 58232,
                     25362, 24622],
       'moons': [0, 0, 1, 2, 80, 83, 27, 14]
        }
df = pd.DataFrame(data)
df
```

```python
print(df['moons']<20)
```

```python
print(df[df['moons']<20])
```

```python
mask = df['moons']<20
df[mask]
```

```python
mask = (df['moons']<10) | (df['moons']>50)
print(mask)
```

```python
mask = (df['moons']<10) | (df['moons']>50)
df[mask]                                       # it will give dataframe of true conditions only
```

```python
mask = (df['moons']>20) & ~(df['moons'] == 80) & ~(df['radius_km']<5000)
df[mask]
```

```python
mask = (df['moons']>20) & (df['moons'] != 80) & (df['radius_km']>5000)         # same statement different syntax just to understand
df[mask]
```
