There are several different ways to import a CSV file into Python, but we will only review some of the more common ways. Start by using a with statement and open() function.  Pass the file name (or file path) of the CSV file to the open() function along with an argument for the mode parameter of the function.  

with open("file path/file name", mode=)

Typically, you’ll be defining the mode inside the with open() argument field as 'r' because you want Python to open and read the CSV file.

Next, we’ll add as file to the end, which is assigning the result to a variable name. In this case, we’ll name it data.

```python
with open('/content/j-CUaNWRQw-glGjVkdMP-w_0399ce5107564abeb17e3bba3a5ab1f1_Modified_Unicorn_Companies.csv', mode='r') as file:
  data = file.read()
```

In above code date is an object. for us to work with the date we need to transform object into ints.


```python
df['date'].dtypes
```

```python
df['date']= pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
df['date'].dtype
```

```python
df.head(10)
```

```python
df['month'] = df['date'].dt.month
df.head()
```

```python
# create a new 'month_nxt' column with month_name with first 3 letters
df['month_txt'] = df['date'].dt.month_name().str.slice(stop=3)
df.head()
```

```python
# create a new helper dataframe for plotting
df_by_month = df.groupby(['month','month_txt']).sum().sort_values('month', ascending=True).head(12).reset_index()
df_by_month
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(x=df_by_month['month_txt'], height=df_by_month['number_of_strikes'], label='Number_of_strikes')

plt.xlabel('months(2018)')
plt.ylabel('Number of lightening strikes')
plt.title('Number of lightening strikes in 2018 by months')

plt.legend()
plt.show()
```
